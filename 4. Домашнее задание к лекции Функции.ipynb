```python
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}
##################################### Funkcia - poisk po nomery dokumenta (p)
def poisk_p (nomer_doka):
    x=0
    for poisk in documents:
        if poisk['number']==nomer_doka:
            print('Владелец документа: ', poisk['name'])
            break
        elif poisk['number']!=nomer_doka:
            x=x+1
    if x==summa_doc:
        print('Документ не найден в базе')

##################################### Funkcia - pokaz nomera polki (s)
def poisk_s (nomer_doka):
    x=0
    for key, value in directories.items():
        if (str(nomer_doka)) in value:
            print('Документ хранится на полке: ', key)
            break
        else:
            x=x+1
    if x==summa_doc:
        print('Документ не найден в базе')

##################################### Funkcia - pokaz vsei infi (i)
def poisk_i (nomer_doka):
    for tipag in documents:
        tip=tipag['type']             # Tip dokumenta
        vladelec=tipag['name']        # Vladelec dokumenta
    for key, value in directories.items():
        for param in value:
            nomer=param                   # Nomer dokumenta
            polka=key                     # Polka dokumenta
            print('№: ',nomer,', ','тип: ',tip,', ','владелец: ',vladelec,', ','полка: ',polka, sep='')

##################################### Funkcia - dobavit novuyu polku
def poisk_ads (nomer_doka):
    nomka=input('Введите номер полки: ')
    if nomka in directories.keys():
        key=list(directories.keys())
        print('Такая полка уже существует. Текущий перечень полок: ', (", ".join(key)), sep='')
    else:
        directories.setdefault(nomka,[])
        key=list(directories.keys())
        print('Полка добавлена. Текущий перечень полок: ', (", ".join(key)), sep='')

##################################### Funkcia - ydalit polku
def poisk_ds (nomer_doka):
    nomka=input('Введите номер полки: ')
    for key, value in list(directories.items()):
        if nomka==key and value==list():
            del(directories[key])
            klu=list(directories.keys())
            itog='Полка удалена. Текущий перечень полок: ', (", ".join(klu))
        elif nomka==key:
            klu=list(directories.keys())
            itog='На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: ', (", ".join(klu))
        elif nomka not in directories.keys():
            klu=list(directories.keys())
            itog='Такой полки не существует. Текущий перечень полок: ', (", ".join(klu))
    print(''.join(itog))

##################################### Telo programmi
komand=('')
summa_doc=len(documents)

while komand != 'q':
    komand=input('Введите команду: ')
    if komand=='p':
        nomer_doka=input('Введите номер документа: ')
        poisk_p(nomer_doka)
    elif komand=="s":
        nomer_doka=input('Введите номер документа: ')
        poisk_s(nomer_doka)
    elif komand=="i":
        poisk_i(nomer_doka)
    elif komand=="ads":
        poisk_ads(nomer_doka)
    elif komand=="ds":
        poisk_ds(nomer_doka)
```
